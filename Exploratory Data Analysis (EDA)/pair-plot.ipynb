```python
# importing all the required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# reading csv file in pandas
url = 'https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv'
iris_data = pd.read_csv(url)
```

# Pair plot 
In our iris dataset there are four features "sepal_length", "sepal_width", "petal_length", "petal_width" and we will construct a pair plot using two features "sepal_length", "sepal_width"

Thus the total number of different pair plots that we can form will be $^4C_2 = 6$ plots. The possible combinations are:
1. "sepal_length" vs "sepal_width"
2. "sepal_length" vs "petal_length"
3. "sepal_length" vs "petal_width"
4. "sepal_width" vs "petal_length"
5. "sepal_width" vs "petal_width"
6. "petal_length" vs "petal_width"

```python
plt.close()
sns.set_style("whitegrid")
sns.pairplot(iris_data, hue = "species", height = 4)
plt.show()
```

As we can see above the graph between "petal_length" and "petal_width" is the best plot to distinguish the iris flower species. Also, we can see a clear seperation for sentosa flower but for versicolor and virginica there is some overlap.

Pair plots are fairly easy to understand but only when the dimentionality of data is less (generally below 6 features). For large dimentional data PCA and t-SNE are more usefull.
