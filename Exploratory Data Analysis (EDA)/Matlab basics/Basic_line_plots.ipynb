```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.array([1,2,3,4])
b = np.array([1,4,9,16])

plt.plot(a,b)

# for title
plt.title("x vs x-square")
plt.xlabel("integer")
plt.ylabel("square of integer")
```

More modifications to the plot for including the different color

```python
plt.plot(a,b, color = "green", linewidth = 2)
```
