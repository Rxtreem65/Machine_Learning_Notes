```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [i for i in range(-3,10)]
```

```python
fig, ax = plt.subplots(figsize = (12,6))

ax.plot(x, x+1, color = 'r', linewidth = 0.25)
ax.plot(x, x+2, color = 'r', linewidth = 0.50)
ax.plot(x, x+3, color = 'r', linewidth = 1)
ax.plot(x, x+4, color = 'r', linewidth = 1.25)

# possible line space options
ax.plot(x, x+5, color = 'r', lw = 3, linestyle = '-')
ax.plot(x, x+6, color = 'r', lw = 3, ls = '-.')
ax.plot(x, x+7, color = 'r', lw = 3, ls = ':')

# custom dash
line, ax = ax.plot(x, x+8, color = 'black', lw= 1.5)
line.set_dashes([5,10,15,10])

# possible marker symbols
ax.plot(x, x+9, color = 'b', lw= 3, ls = "-", marker = '+')
ax.plot(x, x+10, color = 'b', lw= 3, ls = "-_", marker = 'o')
ax.plot(x, x+11, color = 'b', lw= 3, ls = "-", marker = 's')
ax.plot(x, x+12, color = 'b', lw= 3, ls = "--", marker = '1')

# marker size and color
ax.plot(x, x+13, color = 'b', lw= 3, ls = "-", marker = '+')
ax.plot(x, x+14, color = 'b', lw= 3, ls = "-", marker = '+')
ax.plot(x, x+15, color = 'b', lw= 3, ls = "-", marker = '+')
ax.plot(x, x+16, color = 'b', lw= 3, ls = "-", marker = '+')
```

# Legends in graphs

```python
# customising the axes front and color
x = np.array([x*np.pi/16 for x in range(-100,100)])

fig = plt.figure()
ax1 = fig.add_axes([0,1,1,1])
ax1.plot(x, np.sin(x),'b',label = 'sin(x)')
ax1.plot(x, np.cos(x), marker = 'o', color = 'r', label='cos(x)')

ax1.set_xlabel('x', fontsize = 15, color = 'r')
ax1.set_ylabel('sin(x)', fontsize = 20, color = 'b')
ax1.set_title('sin(x) graph', fontsize = 20, color = 'r')

ax1.legend(loc = 3)
```
