# Dealing with sin() and cosin() functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# customising the axes front and color
x = np.array(range(int(-np.pi),int(np.pi)))

fig = plt.figure()
ax1 = fig.add_axes([0,1,1,1])
ax1.plot(x, np.sin(x),'b')

ax1.set_xlabel('x', fontsize = 15, color = 'r')
ax1.set_ylabel('sin(x)', fontsize = 20, color = 'b')
ax1.set_title('sin(x) graph', fontsize = 20, color = 'r')
```

```python
# customising the axes front and color
x = np.array([x*np.pi/16 for x in range(-100,100)])

fig = plt.figure()
ax1 = fig.add_axes([0,1,1,1])
ax1.plot(x, np.cos(x),'ro')

ax1.set_xlabel('x', fontsize = 15, color = 'r')
ax1.set_ylabel('sin(x)', fontsize = 20, color = 'b')
ax1.set_title('sin(x) graph', fontsize = 20, color = 'r')
```
