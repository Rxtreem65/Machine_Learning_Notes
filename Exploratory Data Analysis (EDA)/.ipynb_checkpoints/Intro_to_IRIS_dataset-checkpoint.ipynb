```python
# importing all the required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

We download iris data from the url and read it in iris_data with read_csv(url)

```python
# reading csv file in pandas
url = 'https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv'
iris_data = pd.read_csv(url)
```

We can study the dimentions of data by shape, columns and value_counts

```python
# printing the dimensions of the dataset
print("The shape of the iris dataset:\n",iris_data.shape,"\n")
print("The columns of the iris dataset:\n",iris_data.columns,"\n")
print("The number of species in the datasets with the total count:\n", iris_data["species"].value_counts())
```

As we can see above, we have equal number of datapoints of each species. Thus, Iris dataset is a balanced dataset as it has 50 datapoints of each species. But in real world the datasets are imbalance i.e. the dataset does not have equal number of datapoints for a particular class or label 

### 1. Understanding dataset by scatter plot
Now we plot the iris dataset using plot function with kind = "scatter", x-axis = "sepal_length" and y-axis = "sepal_width" as prameters. Then we use show() from matplot to show the plot

```python
# plotting the iris dataset with plot
iris_data.plot(kind = 'scatter', x = 'sepal_length', y = 'sepal_width')
plt.show()
```

We now see how to plot a dataset using seaborn. Here, hue is responsible for color coding the data based on a set column. In the below code we have diffferent species represented by different color. height represents the height of the plot we wish to plot, in this case it is 4. And then we map the data onto two axis of "sepal_length" and "sepal_width". Finally, we add legends to our plot with add_legend() function

```python
# plotting the iris dataset with seaborn
sns.set_style("whitegrid");
face_grid = sns.FacetGrid(iris_data, 
             hue = "species",
             height = 4).map(plt.scatter,
                       "sepal_length",
                       "sepal_width").add_legend();
plt.show()
```

Now from above plot, we can easily distinguishe the difffrent species and also understand how to seperate classify them based on sepal_width and sepal_length. Also, we can see that distinguishing setosa is easy, but diffentiating versicolor and virginica is difficult.

### 2. Plotting data onto 3 - Dimensions 

```python
from mpl_toolkits.mplot3d import Axes3D
sns.set(style = "darkgrid")

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

# Creating axis for the plot
x = iris_data['sepal_length']
y = iris_data['sepal_width']
z = iris_data['petal_length']
c = iris_data['species']

ax.set_xlabel("sepal_length")
ax.set_ylabel("sepal_width")
ax.set_zlabel("petal_length")

ax.scatter(x, y, z)

plt.show()
```

```python
# Creating figure
fig = plt.figure(figsize = (16, 9))
ax = plt.axes(projection ="3d")
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.3,
        alpha = 0.2)
 
# Creating color map
my_cmap = plt.get_cmap('hsv')
 
# Creating plot
sctt = ax.scatter3D(x, y, z,
                    alpha = 0.8,
                    c = (x + y + z),
                    cmap = my_cmap,
                    marker ='o')
 
plt.title("simple 3D scatter plot")
ax.set_xlabel('X-axis', fontweight ='bold')
ax.set_ylabel('Y-axis', fontweight ='bold')
ax.set_zlabel('Z-axis', fontweight ='bold')
fig.colorbar(sctt, ax = ax, shrink = 0.5, aspect = 5)
 
# show plot
plt.show()
```
